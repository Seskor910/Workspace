```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
from scipy.signal import spectrogram
```

```python
# Load mat files
file_path1 = r'E:\Vscode\Tugas Anime\Workspace\Dataset\EEG_128channels_resting_lanzhou_2015\Depression\file mat\02010002rest 20150416 1017..mat'
mat_data1 = scipy.io.loadmat(file_path1)

# Print all variables
print(mat_data1.keys())

# Print specific variable
variable_data1 = mat_data1['a02010002rest_20150416_1017mat']
print(variable_data1)

eeg_data = variable_data1

# Perhitungan spektogram
frequencies, times, spectrogram_data = spectrogram(eeg_data, fs=1000)

# Pemilihan salah satu channel
# cari paper rata-rata pakai channel berapa untuk depresi
channel_data = spectrogram_data[0]

# Penyesuaian dimensi
channel_data = 10 * np.log10(channel_data[:-1, :-1])

# Visualisasi
plt.pcolormesh(times, frequencies, channel_data, shading='auto')
plt.ylabel('Frequency (Hz)')
plt.xlabel('Time (s)')
plt.title('EEG Spectrogram')
plt.colorbar(label='Power/Frequency (dB/Hz)')
plt.show()
```
